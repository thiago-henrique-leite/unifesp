```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# KMeans e Single Linkage
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
# Carregando os datasets
dataset_aggregation = pd.read_csv("../input/unifesp-2021-01-ia/Dataset_Aggregation.csv")
dataset_d31 = pd.read_csv("../input/unifesp-2021-01-ia/Dataset_D31.csv")
dataset_flame = pd.read_csv("../input/unifesp-2021-01-ia/Dataset_Flame.csv")
dataset_pathbased = pd.read_csv("../input/unifesp-2021-01-ia/Dataset_Pathbased.csv")
```

## **Dataset Agreggation**

```python
# Transforma as linhas da planilha em um array
X = np.array(dataset_aggregation) 

# Calcula quantidade de clusters
n_clusters = int(max(dataset_aggregation['Z']))

# Plota o grfico dos grupos com base nos dados fornecidos no dataset
plt.scatter(X[:, 0], X[:, 1], c=np.array(dataset_aggregation['Z']), s=50,  cmap='viridis')
```

- ### **KMeans**

```python
# Realiza a diviso dos clusters pelo algoritmo KMeans
clf_kmeans = KMeans(n_clusters=n_clusters, random_state=0)
clf_kmeans.fit(X)

labels = clf_kmeans.labels_

# Plota o grfico dos clusters encontrados
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

- ### **Single Linkage**

```python
# Realiza a diviso dos clusters pelo algoritmo Single Linkage
clf_single_linkage = AgglomerativeClustering(n_clusters = n_clusters, linkage = 'single')
clf_single_linkage.fit(X)

labels = clf_single_linkage.labels_

# Plota o grfico dos clusters encontrados
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
```

- ### **Resultado**

#### Para o seguinte dataset, os dois algoritmos no tiveram bons resultados, apesar de gerarem o mesmo esqueleto, e ambos bem parecidos com o esqueleto do grfico esperados, os grupos esto diferentes. Visualmente, analisando os dois grficos gerados, acredito que o Single Linkage teve uma pequena vantagem.
